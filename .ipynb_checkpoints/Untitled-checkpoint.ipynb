## Business Understanding
### Project Overview

The aim of this project is to develop a binary classification model that predicts whether an arrest will occur after a Terry Stop. This analysis aims to provide insights into the factors influencing stop outcomes and explore the potential impact of features such as the presence of weapons, time of day, officer characteristics (gender, race), and subject characteristics (perceived race, perceived gender). This algorithm, as well as the insights generated from this study will be crucial for law enforcement agencies optimize their procedures during Terry Stops, leading to more informed decision-making and potentially improving overall public safety. The project will be conducted with a focus on transparency, fairness, and ethical considerations.

### Objectives
* Build a machine learning model capable of binary classification to predict whether an arrest will occur after a Terry Stop.
* Explore the impact of the presence of weapons on the likelihood of arrest during Terry Stops.
* Investigate whether officer characteristics, including gender and race, play a role in stop outcomes.
* Investigate whether time of day has an impact on the Terry stops.
* Transparently communicate the use of sensitive attributes, specifically race and gender, in the model development process.

## Data Understanding
This data represents records of police reported stops under Terry v. Ohio, a landmark Supreme Court case in 1967-8. This is When a police officer has a reasonable suspicion that an individual is armed, engaged in, or about to be engaged in criminal conduct, the officer may briefly stop and detain an individual for a frisk or search.
Each record contains perceived demographics of the subject, as reported by the officer making the stop and officer demographics as reported to the Seattle Police Department. 

### Data Structure

2. Number of columns = 23 
3. The data format is CSV. 

### Column names and their descriptions
The following are the descriptions of the columns in the dataset.
<ul>
<li><b>Subject Age Group:</b> The Age Group of the subject(10 year increments) as reported by the officer.</li>
<li><b>Subject ID:</b> A unique identifier for the subject.</li>
<li><b>GO / SC Num:</b> General Offense or Street Check number, relating the Terry Stop to the parent report.</li>
<li><b>Terry Stop ID:</b> Key identifying unique Terry Stop reports.</li>
<li><b>Stop Resolution:</b> Resolution of the stop as reported by the officer.</li>
<li><b>Weapon Type:</b> Type of weapon, if any, identified during a search or frisk of the subject. Indicates "None" if no weapons was found.</li>
<li><b>Officer ID:</b> An identifier for the officer involved in the Terry Stop.</li>
<li><b>Officer YOB:</b> Year of birth, as reported by the officer.</li>
<li><b>Officer Gender:</b> Gender of the officer, as reported by the officer.</li>
<li><b>Officer Race:</b> The race of the officer involved in the Terry Stop.</li>
<li><b>Subject Perceived Race:</b> The race perceived by the officer for the subject involved in the Terry Stop.</li>
<li><b>Subject Perceived Gender:</b> The gender perceived by the officer for the subject involved in the Terry Stop.</li>
<li><b>Reported Date:</b> The date when the Terry Stop was reported.</li>
<li><b>Reported Time:</b> Time the stop was reported in the Records Management System (RMS). Not the time the stop occurred but generally within 10 hours.</li>
<li><b>Initial Call Type:</b> Initial classification of the call as assigned by 911.</li>
<li><b>Final Call Type:</b> Final classification of the call as assigned by the primary officer closing the event.</li>
<li><b>Call Type:</b> How the call was received by the communication center.</li>
<li><b>Officer Squad:</b> Functional squad assignment (not budget) of the officer as reported by the Data Analytics Platform (DAP).</li>
<li><b>Arrest Flag:</b> Indicator of whether a "physical arrest" was made, of the subject, during the Terry Stop. Does not necessarily reflect a report of an arrest in the Records Management System (RMS).</li>
<li><b>Frisk Flag:</b> Indicator of whether a "frisk" was conducted, by the officer, of the subject, during the Terry Stop.</li>
<li><b>Precinct:</b> Precinct of the address associated with the underlying Computer Aided Dispatch (CAD) event. Not necessarily where the Terry Stop occurred.</li>
<li><b>Sector:</b> Sector of the address associated with the underlying Computer Aided Dispatch (CAD) event. Not necessarily where the Terry Stop occurred.</li>
<li><b>Beat:</b> Beat of the address associated with the underlying Computer Aided Dispatch (CAD) event. Not necessarily where the Terry Stop occurred.</li>
</ul>

## Feature Engineering

```python
# Convert the 'Reported Date' column to datetime format
df['Reported Date'] = pd.to_datetime(df['Reported Date'])

# Extract date components
df['Reported Day'] = df['Reported Date'].dt.day
df['Reported Month'] = df['Reported Date'].dt.month
df['Reported Year'] = df['Reported Date'].dt.year

# Display the resulting DataFrame
df.head()
```

```python
# Convert the 'Reported Time' column to datetime format
df['Reported Time'] = pd.to_datetime(df['Reported Time'], format='%H:%M:%S.%f')

# Extract time components 
df['Reported Hour'] = df['Reported Time'].dt.hour
df['Reported Minute'] = df['Reported Time'].dt.minute

# Display the resulting DataFrame
df.head()
```

```python
#Function to categorize the time of day by the hours
def categorize_time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 24:
        return 'Evening'
    else:
        return 'Night'

df['Time of Day'] = df['Reported Hour'].apply(categorize_time_of_day)
```

### Check presence of weapons on the likelihood of arrest during Terry Stops

```python
# Extract rows where no weapon was involved:
no_weapon = df[df['Weapon Type'] == 'None']

# Calculate the percentage of cases without weapons compared to total terry stops per age group:
no_weapon_percentage = (no_weapon['Weapon Type'].groupby(df['Subject Age Group']).count() /
                        df['Weapon Type'].groupby(df['Subject Age Group']).count()) * 100

# Reorder the percentages to align with age ranges:
no_weapon_percentage = no_weapon_percentage.reindex(["1 - 17", "18 - 25", "26 - 35", "36 - 45", "46 - 55",
                                                     "56 and Above", "-"])
# Convert to a DataFrame:
no_weapon_df = pd.DataFrame(no_weapon_percentage)

# Extract rows where a firearm was present:
firearm = df[df['Weapon Type'] == 'Firearm']
firearm_percentage = (firearm['Weapon Type'].groupby(df['Subject Age Group']).count() /
                      df['Weapon Type'].groupby(df['Subject Age Group']).count()) * 100
firearm_percentage = firearm_percentage.reindex(["1 - 17", "18 - 25", "26 - 35", "36 - 45", "46 - 55",
                                                 "56 and Above", "-"])
firearm_df = pd.DataFrame(firearm_percentage)

# Extract rows where a non-firearm weapon was present:
non_firearm = df[df['Weapon Type'] == 'Non-Firearm']
non_firearm_percentage = (non_firearm['Weapon Type'].groupby(df['Subject Age Group']).count() /
                          df['Weapon Type'].groupby(df['Subject Age Group']).count()) * 100
non_firearm_percentage = non_firearm_percentage.reindex(["1 - 17", "18 - 25", "26 - 35", "36 - 45", "46 - 55",
                                                         "56 and Above", "-"])
non_firearm_df = pd.DataFrame(non_firearm_percentage)

# Combine the three DataFrames into one and reset the index:
combined_weapon = pd.concat([no_weapon_df, non_firearm_df, firearm_df], axis=1)
combined_weapon.reset_index(inplace=True)

# Create a horizontal stacked bar chart comparing the percentage of weapon types per subject age group:
combined_weapon.plot(x='Subject Age Group', y="Weapon Type", kind="barh", stacked=True, figsize=(16, 10))
plt.gca().invert_yaxis()  # Reverse the order of the y-axis so 1-17 is at the top

# Add chart title, labels, and legend
plt.title('% of Weapon Type Per Subject Age Group', fontsize=20)
plt.xlabel('% of Weapon Type', fontsize=14)
plt.ylabel('Subject Age Group', fontsize=14)
plt.legend(labels=['No Weapon', 'Non-Firearm', 'Firearm'])
plt.show()
```

### checking significance

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['Officer Gender'], df['Stop Resolution'])

# Perform the chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-Square Value: {chi2}")
print(f"P-value: {p}")

# Check the significance level
alpha = 0.05
if p < alpha:
    print("There is a significant association between Officer Gender and Arrest.")
else:
    print("There is no significant association between Officer Gender and Arrest.")
```

```python
# Define a function to plot a confusion matrix:
def confusion_matrix_plot(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    # Function to create a confusion matrix chart for model performance visualization
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j], horizontalalignment='center', color='white' if cm[i, j] > thresh else 'black')
    plt.colorbar()
    plt.show()
```

### K-Nearest Neighbors Classification

```python
# Determine the optimal k value for KNN classification:
from sklearn.neighbors import KNeighborsClassifier

def find_best_k(X_train, y_train, X_test, y_test, min_k=1, max_k=25):
    best_k = 0
    best_score = 0.0
    for k in range(min_k, max_k + 1, 2):
        knn = KNeighborsClassifier(n_neighbors=k, algorithm='brute')
        knn.fit(X_train, y_train)
        preds = knn.predict(X_test)
        accuracy = accuracy_score(y_test, preds)
        if accuracy > best_score:
            best_k = k
            best_score = accuracy
    
    print("Best Value for k: {}".format(best_k))
    print("Accuracy Score: {:.4f}".format(best_score))
```

```python
# Call the function to find the best k value:
find_best_k(X_train_ohe, y_train, X_test_ohe, y_test)
```

```python
# Create the KNN classifier with the best k value:
knn = KNeighborsClassifier(n_neighbors=25, algorithm='brute')
knn.fit(X_train_ohe, y_train)
y_pred_knn = knn.predict(X_test_ohe)
# Check the accuracy of the KNN model:
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print('KNN Accuracy: {:.2f}%'.format(accuracy_knn * 100))
# Print the classification report for KNN:
print(classification_report(y_test, y_pred_knn))
```

```python
# Create the confusion matrix for KNN:
cm_knn = confusion_matrix(y_test, y_pred_knn)

# Plot the confusion matrix:
confusion_matrix_plot(cm_knn, classes=class_names, title='KNN Confusion Matrix')
```
